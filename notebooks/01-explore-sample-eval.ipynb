# Explore Example Evaluation Report

This notebook loads the example evaluation JSON located at `docs/onboarding/example-evaluation-report.json`, converts the `report` object into a pandas DataFrame (one row per analyzer), and produces two simple visualizations:

- Bar chart of per-analyzer scores
- Histogram of analyzer scores

Use this as a starting point for more advanced EDA or to adapt for bulk-conversion scripts.

```python
# Cell 1: Load libraries and the example JSON
import json
from pathlib import Path
import pandas as pd

# Path to the example JSON included in the repo
p = Path('docs/onboarding/example-evaluation-report.json')
obj = json.loads(p.read_text())

# Quick inspect
print('Run id:', obj.get('id'))
print('Status :', obj.get('status'))
print('Contextual debt score:', obj.get('contextualDebtScore'))

# Convert report -> DataFrame (one row per analyzer)
report = obj.get('report', {})
rows = []
for k, v in report.items():
    rows.append({
        'analyzer': k,
        'score': v.get('score'),
        'details': v.get('details')
    })
df = pd.DataFrame(rows)
df
: 
,
: {
: 

: [
2
4
,
,
1
,
,
,
,
,
,
,
,
,
3
,
,
,
,
,
,

: 
,
: {
: 

: [
,
,
,
,
,
```
